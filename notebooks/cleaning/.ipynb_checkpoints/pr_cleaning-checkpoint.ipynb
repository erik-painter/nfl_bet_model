# Player Rating Cleaning

Will conduct some neccessary cleaning of player ratings data. 

```python
import numpy as np
import pandas as pd
```

```python
df_20 = pd.read_csv('/Users/epainter/Desktop/bet_model_v2/data/raw/pr_2020_raw.csv')
df_21 = pd.read_csv('/Users/epainter/Desktop/bet_model_v2/data/raw/pr_2021_raw.csv')
df_22 = pd.read_csv('/Users/epainter/Desktop/bet_model_v2/data/raw/pr_2022_raw.csv')
df_23 = pd.read_csv('/Users/epainter/Desktop/bet_model_v2/data/raw/pr_2023_raw.csv')
```

I will first filter out the *Name* column as it has no effect for our analysis.

```python
df_list = [df_20, df_21, df_22, df_23]


for df in df_list:
    
    df.drop(columns = ['Name'], inplace = True)   
```

## Merging Player Ratings

Now that every data frame for each NFL season is cleaned, I will now merge them into a single data frame so that
it can be used for easier analysis later.

Since each team has multiple players at some positions (and not every team has the same amount of players at each position) I will first pivot the tables by creating a column for each position and the value for each column will be the average of the players at that position for a given team if the number of players at a given position is greater than zero.

```python
# Player rankings are the mean at each position if number of players at a position is > 1

df_20_pivot = df_20.pivot_table(index='Team', columns='Position', values='Rating', 
                                            aggfunc=lambda x: round(x.mean(), 2))

df_21_pivot = df_21.pivot_table(index='Team', columns='Position', values='Rating', 
                                            aggfunc=lambda x: round(x.mean(), 2))

df_22_pivot = df_22.pivot_table(index='Team', columns='Position', values='Rating', 
                                            aggfunc=lambda x: round(x.mean(), 2))

df_23_pivot = df_23.pivot_table(index='Team', columns='Position', values='Rating', 
                                            aggfunc=lambda x: round(x.mean(), 2))

# Have to create a column that specifies the season

df_20_pivot['Season'] = 2020
df_21_pivot['Season'] = 2021
df_22_pivot['Season'] = 2022
df_23_pivot['Season'] = 2023

# Concatenate ranking dfs for different years into a single df -> rating_df

rating_df = pd.concat([df_20_pivot, df_21_pivot, df_22_pivot, df_23_pivot], ignore_index=False)
rating_df
```

The data is now cleaned!

# Write to CSV

```python
fp = '/Users/epainter/Desktop/bet_model_v2/data/clean/pr_clean.csv'

rating_df.to_csv(fp, index=True)

print(f"Data saved to {fp}")
```
