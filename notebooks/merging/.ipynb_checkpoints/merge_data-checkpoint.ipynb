# Merging All Cleaned Data

Here I will be merging all of the data frames that we have cleaned. This will allow us to have one 'master' data frame where we can conduct analysis, perform feature engineering, and ultimately refine and process so that we can create and run our model.

```python
import os
import numpy as np
import pandas as pd
```

```python
pd.set_option('display.max_columns', None)
```

```python
bet_df = pd.read_csv('/Users/epainter/Desktop/bet_model_v2/data/clean/hist_bet_clean.csv')
pr_df = pd.read_csv('/Users/epainter/Desktop/bet_model_v2/data/clean/pr_clean.csv')
gd_df = pd.read_csv('/Users/epainter/Desktop/bet_model_v2/data/clean/gd_clean.csv')
teams_df = pd.read_csv('/Users/epainter/Desktop/bet_model_v2/data/clean/nfl_teams.csv')
```

```python
pr_df
```

# 1.) Merging Player Ratings onto Historical Bet Data

The first step is to merge the player ratings onto the historical betting data frame. There is no reason to merge these first, I just happened to do it this way.

It will merge the player ratings for both the favorite and underdog, where it will specify with either a *_fav* or *_und* suffix.

```python
# Merge bet_df with both favorite and underdog roster ratings
merged_df = bet_df.merge(pr_df, how='left', left_on=['season', 'favorite'], 
                         right_on=['Season', 'Team']).merge(pr_df, how='left', 
                                                                 left_on=['season', 'underdog'], 
                                                                 right_on=['Season', 'Team'], 
                                                                 suffixes=('_fav', '_und'))
```

```python
merged_df.head()
```

```python
pr_df[(pr_df['Team'] == 'Houston Texans') & (pr_df['Season'] == 2020)]
```

We successfully merged the data frames, but now there are some extra columns we can get rid of.

```python
cols_to_drop = ['Team_fav', 'Season_fav', 'Team_und', 'Season_und']

merged_df.drop(columns=cols_to_drop, inplace=True)
```

# 2.) Merging Game by Game Data 

Here we will be merging our game by game data onto our working merged data frame. However, there are some issues to address before just going and merging them like we did for the player ratings data.

Since our master data frame really goes off teams by who is the favorite or underdog and our game data frame goes by winner and loser, we need to create columns that specify the winner and loser for a given game.

This will allow us to merge the data correctly.

```python
# will use np.where

# Winner
merged_df['winner'] = np.where(merged_df['fav_score'] > merged_df['und_score'], merged_df['favorite'], merged_df['underdog'])

# Loser
merged_df['loser'] = np.where(merged_df['fav_score'] > merged_df['und_score'], merged_df['underdog'], merged_df['favorite'])

# This is to check whether it specifies the correct winner/loser for a given game
merged_df[['winner', 'loser', 'favorite', 'underdog', 'fav_score', 'und_score']]
```

Now that we have a column that specifies the winner and loser for any given game, we can merge the data frames.

This will provide the stats in terms of winner or loser, which we will address in later during the feature engineering phase.

There is one more issue, I noticed that in the game by game data frame, there are team names for both the *Washington Commanders* and the *Washington Football Team*. I need to replace *Washington Football Team* with *Washington Commanders* so we don't have merging issues later.

```python
gd_df['winner'] = gd_df['winner'].replace('Washington Football Team', 'Washington Commanders')

gd_df['loser'] = gd_df['loser'].replace('Washington Football Team', 'Washington Commanders')
```

```python
# Merge bet_df with both favorite and underdog roster ratings
merged_df = merged_df.merge(gd_df, how='left', left_on=['week', 'season', 'winner'], 
                         right_on=['week', 'season', 'winner'])

merged_df
```

```python
# Now we can drop the unecessary columns

merged_df = merged_df.drop(columns=['loser_y'])

# Want to clean the data to only get necessary information and useful columns

merged_df = merged_df.rename(columns={'loser_x': 'loser'})

merged_df.head()
```

# 3.) Merging Teams

This is a small yet important step. I want to merge data on whether the favorite and underdog are within the same division.

```python

merged_df = merged_df.merge(
    teams_df, how='left', left_on='favorite', right_on='team_name').merge(
    teams_df, how='left', left_on='underdog', right_on='team_name', suffixes=('_fav', '_und'))

merged_df = merged_df.drop(columns=['team_name_fav', 'team_name_short_fav', 'team_id_fav', 
                                            'team_id_pfr_fav', 'team_conference_fav',
                                           'team_name_und', 'team_name_short_und', 'team_id_und', 
                                           'team_id_pfr_und', 'team_conference_und'])
```
