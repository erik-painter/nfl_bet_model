```python
# I want to do some data imputation for some of the lagged features in this model for those columns containing
# NaN values in the first week so that we have some data

# I could not include the first week (week 1 - 2022) in my model, however we would lose data, making the overall 
# model less accurate
```

```python
# Function to impute by group
def impute_by_group(df, group_col, target_col):
    # Calculate mean for each group
    group_means = df.groupby(group_col)[target_col].transform('mean')
    
    # Create a mask for NaN values
    mask = df[target_col].isna()
    
    # Impute NaN values with the group mean
    df.loc[mask, target_col] = group_means[mask]
    
    return df

# Assuming 'fav_team' is the column that indicates favorite/underdog
test_df = impute_by_group(df, 'favorite', 'fav_ppg_lag')

test_df
```

```python
def merging_all_fav_imp_values(df, target_cols):
    
    for col in target_cols:
        
        df = impute_by_group(df, 'favorite', col)
    
    return df
```

```python
df.columns
```

```python
fav_columns_to_imp = ['fav_ppg_lag', 'und_ppg_lag',
       'fav_papg_lag', 'und_papg_lag', 'fav_ypg_lag', 'und_ypg_lag',
       'fav_yapg_lag', 'und_yapg_lag', 'fav_topg_lag', 'und_topg_lag',
       'fav_tofpg_lag', 'und_tofpg_lag', 'fav_avg_mov_lag', 'und_avg_mov_lag',
                      'fav_win_pct_lag', 'und_win_pct_lag',
       'fav_last_5_win_pct_lag', 'und_last_5_win_pct_lag',
       'fav_home_win_pct_lag', 'und_home_win_pct_lag', 'fav_away_win_pct_lag',
       'und_away_win_pct_lag']

test_df = merging_all_fav_imp_values(df, fav_columns_to_imp)

test_df
```

```python
# Get a list of columns that end with '_lag'
lag_columns = [col for col in test_df.columns if col.endswith('_lag')]

# Round these columns to 2 decimal places
for col in lag_columns:
    test_df[col] = test_df[col].round(2)
    
test_df
```
