# 1.) Differences, Sums, and Ratios

I am going to experiment by building various aggregate features. First I want to create features that are simply differences, sums, or ratios between features.

**Note:** These features are made in terms of the favorite. Meaning that they are always used to highlight the difference of the favorites feature compared to the underdogs.

    - For example, the average differnce MOV is the favorites MOV minus the underdogs MOV

```python
# Differences

df['ppg_diff'] = df['fav_ppg_lag'] - df['und_ppg_lag']
df['ypg_diff'] = df['fav_ypg_lag'] - df['und_ypg_lag']
df['topg_diff'] = df['fav_topg_lag'] - df['und_topg_lag']
df['avg_mov_diff'] = df['fav_avg_mov_lag'] - df['und_avg_mov_lag']
df['win_pct_diff'] = df['fav_win_pct_lag'] - df['und_win_pct_lag']
df['last_5_win_pct_diff'] = df['fav_last_5_win_pct_lag'] - df['und_last_5_win_pct_lag']
```

```python
df.head()
```

```python
# Sums

df['team_ovr_diff'] = (df['off_rating_fav'] - df['def_rating_und']) + (df['def_rating_fav'] - df['off_rating_und'])
df['ypg_sum'] = df['fav_ypg_lag'] + df['und_ypg_lag']
```

```python
# Ratios

df['ppg_ratio'] = df['fav_ppg_lag'] / df['und_ppg_lag']
df['ypg_ratio'] = df['fav_ypg_lag'] / df['und_ypg_lag']
df['avg_mov_ratio'] = df['fav_avg_mov_lag'] / df['und_avg_mov_lag']

# Have to round these columns to 2 decimal places
df['ppg_ratio'] = df['ppg_ratio'].round(2)
df['ypg_ratio'] = df['ypg_ratio'].round(2)
df['avg_mov_ratio'] = df['avg_mov_ratio'].round(2)
```
